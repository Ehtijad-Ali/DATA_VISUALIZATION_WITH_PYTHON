```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Superstore-Dataset.csv')
data.head(6)
```

```python
plt.figure(figsize = (5, 4))
category = data.groupby('Category')['Sales'].sum()
category.plot(kind ='bar', color = 'purple', )
plt.title('Totle Sales By Category', color = 'r')
plt.xlabel('Category', color = 'g')
plt.ylabel('Sales', color = 'g')
plt.grid(color = 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 4))

category_sales = data.groupby('Category')['Sales'].sum()

plt.bar(category_sales.index, category_sales.values, color='skyblue')
plt.title('Total Sales By Category', fontsize=18, fontweight='bold')
plt.xlabel('Category', fontsize=16, color='g')
plt.ylabel('Sales', fontsize=16, color='g')
plt.xticks(rotation=45)

for i, value in enumerate(category_sales.values):
    plt.text(i, value + 500, f'{value:.2f}', ha='center')

plt.show()
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize = (5, 4))
sns.barplot(x = 'Category', y = 'Sales', data = data,hue = 'Segment', color = 'r', palette = 'icefire', saturation = 10)
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sub_category =  data.groupby('Sub-Category')['Sales'].sum()
plt.boxplot(sub_category.values, vert=False, patch_artist=True)
plt.title('Sales Distribution By Sub_Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks([1], ['sub_category'])
plt.show()

            
```

```python
plt.figure(figsize=(5, 4))
sub_category =  data.groupby('Sub-Category')['Sales'].sum()
plt.boxplot(sub_category.values, vert=False, patch_artist=True,
         boxprops=dict(facecolor='lightblue', color='darkblue'),
         whiskerprops=dict(color='darkblue'),
          capprops=dict(color='darkblue'), 
)

plt.title('Sales Distribution By Sub_Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks([1], ['sub_category'])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sales_data = data.groupby(['Region', 'Category'])['Sales'].sum().unstack()

sns.heatmap(sales_data, annot=True, fmt=".2f", cmap="gist_heat", linewidths=.5)
plt.title('Sales by Region and Category', fontsize=18, fontweight='bold')
plt.xlabel('Category', fontsize=14)
plt.ylabel('Region', fontsize=14)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 5))
sales_data = data.groupby(['Region', 'Category'])['Sales'].sum().unstack()
sns.heatmap(
    sales_data, 
    annot=True, 
    fmt=".2f", 
    cmap="coolwarm",  
    linewidths=.5, 
    linecolor='white', 
    annot_kws={"size": 12, "weight": "bold"}, 
    cbar_kws={'label': 'Total Sales'} 
)

plt.title('Total Sales by Region and Category', fontsize=20, fontweight='bold', pad=20)
plt.xlabel('Category', fontsize=16, labelpad=15)
plt.ylabel('Region', fontsize=16, labelpad=15)

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 4))
sns.boxplot(x='Ship Mode', y='Profit', data=data, palette="Set3")

plt.title('Distribution of Profits Across Different Ship Modes', fontsize=18, fontweight='bold')
plt.xlabel('Ship Mode', fontsize=14, labelpad=10)
plt.ylabel('Profit', fontsize=14, labelpad=10)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
sns.boxplot(
    x='Ship Mode', 
    y='Profit', 
    data=data, 
    palette="coolwarm",   
    notch=True,          
    linewidth=2         
)

plt.title('Profit Distribution Across Ship Modes', fontsize=20, fontweight='bold', color='darkblue', pad=20)
plt.xlabel('Ship Mode', fontsize=16, labelpad=15, color='darkred')
plt.ylabel('Profit ($)', fontsize=16, labelpad=15, color='darkred')

# Customize tick labels
plt.xticks(fontsize=14, color='darkgreen')
plt.yticks(fontsize=14, color='darkgreen')

plt.grid(True, linestyle='--', linewidth=0.5, color='gray', axis='y')

plt.tight_layout()

plt.show()
```
